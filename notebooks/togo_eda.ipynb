```python
#import libraries we want to use to make the EDA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
filePath = "../data/togo-dapaong_qc.csv"
df = pd.read_csv(filePath)
df.head()
```

```python
#get statistical data for the whole data
df.describe()
```

```python
#each columns non-null value and data types
df.info()
```

```python
#number of null in every column
df.isna().sum()
```

```python
#compute missing ration
df.isna().mean().sort_values(ascending=False)
```

```python
from scipy.stats import zscore

colsToCheck = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

zScores = df[colsToCheck].apply(zscore)
zScores
```

```python
#outlier
outliers = (zScores.abs() > 3).sum()
outliers
```

```python
mask = (zScores.abs() < 3).all(axis = 1)
dfClean = df[mask]
dfClean.describe()
```

```python
#save the clean data
dfClean.to_csv("../data/togo_clean.csv")
```

```python
df = dfClean.copy()

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

plt.figure(figsize=(12, 6))
plt.plot(df['Timestamp'], df['GHI'], label='GHI')
plt.plot(df['Timestamp'], df['DNI'], label='DNI')
plt.plot(df['Timestamp'], df['DHI'], label='DHI')
plt.legend()
plt.title('Solar Irradiance Over Time')
```

```python
#seasonally
df_season = df.copy()
df_season['month'] = df_season['Timestamp'].dt.month
df_season.groupby('month')[['GHI', 'DNI', 'DHI']].mean().plot(kind='bar')
```

```python
#daily
df_daily = df.copy()
df_daily['hour'] = df_daily['Timestamp'].dt.hour
df_daily.groupby('hour')[['GHI', 'DNI', 'DHI']].mean().plot()
```
